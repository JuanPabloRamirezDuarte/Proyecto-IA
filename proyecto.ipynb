#Reconocimiento de incendios forestales a partir de imágenes RGB

##Introducción

###El Problema de los incendios forestales

- Los incendios forestales son una preocupación ecológica real en el mundo:
image.png

- Entre 2013 y 2021 se registró una pérdida de 150.575 hectáreas de bosques en Colombia

image.png

- Sin embargo, Colombia no es el peor caso:
image.png

- Entre junio de 2019 y mayo de 2020, Australia pasó por una temporada de incendios forestales que consumió más de 23.000.000 de hectáreas, más de 2500 edificios (incluidas más de 1.300 viviendas) y quitando la vida a al menos 26 personas. Esta fue una de las peores temporadas de incendios que pasó ese país desde la temporada 1974-75, en la cual se consumieron 117.000.000 hectáreas.

image.png

- ‘’*Según datos revelados por el Instituto Nacional de Investigación Espacial (Inpe), la Amazonía brasileña registró en junio (de 2021) 2.308 focos de incendio, su peor cifra desde 2007, con 3.519. Si se compara este índice con las quemas de junio de 2020, que también marcaron un récord en su momento, el incremento fue del 2,6%.*‘’ (https://www.france24.com/es/am%C3%A9rica-latina/20210703-brasil-amazonia-incendios-jair-bolsonaro)

###Una estrategia que se ha utilzado para el monitoreo forestal ha sido el despliegue de drones equipados con cámaras

"*El Cabildo de Tenerife añadió a la campaña contra incendios forestales tres nuevos drones con dos tipos de cámara incorporadas, una tradicional y otra térmica, que se añaden así a las tareas de control.*" (tomado de: https://diariodeavisos.elespanol.com/2022/07/incendio-en-tenerife-drones/)

*“Los llamados drones pueden monitorear y medir muchas cosas, incluyendo el carbono. Son rápidos, baratos, inmediatos y ahorrarán una enorme cantidad de trabajo pesado”* Michael McCall, investigador senior de la Universidad Nacional Autónoma de México (UNAM) y uno de los autores de un documento sobre el tema: https://www.mdpi.com/1999-4907/5/6/1481

El estudio descubrió que fijar una pequeña cámara a un dron del tamaño de un avión de aeromodelismo y hacerlo volar a una altura de 50 a 300 metros puede producir imágenes de alta resolución espacial. A dicha resolución, se puede identificar y monitorear fácilmente árboles específicos y los cambios existentes en la cobertura vegetal.

“Si bien el equipo fotográfico sigue siendo caro, hay quienes están haciendo pruebas usando tan solo una cámara digital común y corriente y tomando gran cantidad de fotografías. Con un programa informático adecuado, es posible unirlas para crear una imagen de calidad a partir de ellas”, señaló McCall.

Una desventaja de esta técnica es que los drones pequeños se ven restringidos en la cantidad de equipos que pueden llevar. Esto limita la calidad de los sensores de imágenes que se pueden conectar al dron, así como el tamaño de las baterías a bordo, lo cual tiende a reducir el tiempo de vuelo a menos de una hora.

##¿En que consiste nuestro proyecto?

Proponemos una inteligencia artificial con la capacidad de reconocer un incendio forestal a partir de una foto en formato RGB. Esta inteligencia artificial podría adaptarse y ser útil en el monitero de bosques por medio del despliegue de drones, ya que ayudaría a la automatización de la detección de dichos incidentes con costos relativamente bajos tanto en términos computacionales como de dinero.

```python
#Librerias 
import pandas as pd
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
```

###Imágenes para entrenamiento

- Lectura de imágenes

```python
X_train=[] 
y_train=[] 
from skimage.transform import resize
y=0
#carga la imagenes
dir=[
    "/content/drive/MyDrive/poyecto ia/archive (1).zip (Unzipped Files)/Forest Fire_Dataset/train/fire",
    "/content/drive/MyDrive/poyecto ia/archive (1).zip (Unzipped Files)/Forest Fire_Dataset/train/nofire",
     "/content/drive/MyDrive/poyecto ia/archive (1).zip (Unzipped Files)/Forest Fire_Dataset/train/smoke",
     "/content/drive/MyDrive/poyecto ia/archive (1).zip (Unzipped Files)/Forest Fire_Dataset/train/smokefire"
     ]
for path in dir:
  for i in os.listdir(path):
    X_train.append(resize(plt.imread(path+'/'+i),(100,100))) 
    y_train.append(y)
  y=y+1
```

- Se revuelven las imágenes

```python
perm = np.random.permutation(range(len(X_train)))
X_train = np.r_[X_train]
y_train=np.r_[y_train]
X_train=X_train[perm]
y_train=y_train[perm]
print(X_train.shape)
```

- visualización

```python
archivo = ['fire','no fire','smoke','smokefire']
plt.figure(figsize=(15,15))
for i in range(10):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(X_train[i], cmap=plt.cm.Greys_r)
    plt.xlabel(archivo[y_train[i]])
```

###Imágenes para testeo

- lectura de las imágenes

```python
X_test=[] 
y_test=[] 
from skimage.transform import resize
y=0
dir=[
    '/content/drive/MyDrive/poyecto ia/archive (1).zip (Unzipped Files)/Forest Fire_Dataset/test/fire',
     '/content/drive/MyDrive/poyecto ia/archive (1).zip (Unzipped Files)/Forest Fire_Dataset/test/nofire',
     '/content/drive/MyDrive/poyecto ia/archive (1).zip (Unzipped Files)/Forest Fire_Dataset/test/smoke',
     '/content/drive/MyDrive/poyecto ia/archive (1).zip (Unzipped Files)/Forest Fire_Dataset/test/smokefire'
]
#carga la imagenes
for path in dir:
  for i in os.listdir(path):
    X_test.append(resize(plt.imread(path+'/'+i),(100,100))) 
    y_test.append(y)
  y=y+1
X_test=np.array(X_test)
y_test=np.array(y_test)
del y
del dir
```

- Revolver

```python
perm = np.random.permutation(range(len(X_test)))
X_test = np.r_[X_test]
y_test=np.r_[y_test]
X_test=X_test[perm]
y_test=y_test[perm]
print(X_test.shape)
del perm
```

- visualización

```python
plt.figure(figsize=(15,15))
for i in range(10):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(X_test[i], cmap=plt.cm.Greys_r)
    plt.xlabel(archivo[y_test[i]])
```

```python
print("cantidad de imágenes para entrenamiento: ",X_train.shape[0])
print("cantidad de imágenes para testeo: ",y_test.shape[0])
print("Cantidad total de imágenes:",X_train.shape[0]+y_test.shape[0])
print("Porcentaje para entrenamiento: ",X_train.shape[0]/(X_train.shape[0]+y_test.shape[0]))
print("dimensiones de las imágenes: ",X_train[0].shape)
```

###Normalización

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train = X_train / 255.
X_test = X_test / 255.
```

##Arquitectura

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.utils import to_categorical
```

```python
train_Y_one_hot = to_categorical(y_train)
test_Y_one_hot = to_categorical(y_test)
 
print('Original label:', y_train[0])
print('After conversion to one-hot:', train_Y_one_hot[0])
```

```python
ker=(5, 5)
model = keras.models.Sequential([
  tf.keras.layers.Conv2D(64, ker, activation='relu', padding='same', input_shape=(100, 100, 3)), 
  tf.keras.layers.Conv2D(32, ker, activation='relu', padding='same'),
  tf.keras.layers.Conv2D(16, ker, activation='relu', padding='same'),
  tf.keras.layers.Conv2D(4, ker, activation='relu', padding='same'),
  tf.keras.layers.MaxPooling2D((2, 2)),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dense(64, activation='relu'),
  tf.keras.layers.Dense(32, activation='relu'),
  tf.keras.layers.Dense(4, activation='softmax')
])
model.summary()
del ker
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

###Entrenamiento

```python
history = model.fit(X_train, y_train, epochs=20, batch_size=70, verbose=1, validation_data=(X_test, y_test))
```

```python
test_loss,test_acc = model.evaluate(X_test, y_test,verbose=1)
print('Test accuracy:', test_acc)
```

```python
accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(len(accuracy))
plt.plot(epochs, accuracy, 'bo', label='Accuracy de entrenamiento')
plt.plot(epochs, val_accuracy, 'b', label='Accuracy de validacion')
plt.title('Accuracy de entrenamiento y validación')
plt.legend()
plt.figure()
plt.plot(epochs, loss, 'bo', label='loss de entrenammiento')
plt.plot(epochs, val_loss, 'b', label='loss de validacion')
plt.title('loss de entrenamiento y validacion')
plt.legend()
plt.show()
```

##Evaluacion del modelo

```python
predicted_classes2 = model.predict(X_test)
```

```python
predicted_classes=[]
for predicted in predicted_classes2:
    predicted_classes.append(predicted.tolist().index(max(predicted)))
predicted_classes=np.array(predicted_classes)
```

```python
predicted_classes.shape, y_test.shape
```

```python
correct = np.where(predicted_classes==y_test)[0]
print("Se encontraron %d etiquetas correctas" % len(correct))
for i, correct in enumerate(correct[0:9]):
    plt.subplot(3,3,i+1)
    plt.imshow(X_test[correct].reshape(500,500), cmap='gray', interpolation='none')
    plt.title("{}, {}".format(predicted_classes[correct], y_test[correct]))
    plt.tight_layout()
```

```python
incorrect = np.where(predicted_classes!=y_test)[0]
print("Se encontraron %d etiquetas incorrectas" % len(incorrect))
for i, incorrect in enumerate(incorrect[0:9]):
    plt.subplot(3,3,i+1)
    plt.imshow(X_test[incorrect].reshape(500,500), cmap='gray', interpolation='none')
    plt.title("{}, {}".format(predicted_classes[incorrect], y_test[incorrect]))
    plt.tight_layout()
```
